```python
import numpy as np
import math
```

### part a

```python
p = []
p.append([0,0,0,0,0,1])

for i in range(1,6):
    # print('\n',i)
    r = []
    p_kk = 0.9  # 1-0.1

    for k in range(i,1,-1):
        p_ik = math.comb(i,k) * 0.1**k * 0.9**(i-k) # probability k out of i machines fail, p(i,i-k)
        # print(f"{i} to {i-k} machines working: {p_ik}")
        p_kk -= p_ik    
        r.append(round(p_ik, 5))  

    r.append(0.1)   # p(i,i-1)
    r.append(p_kk)  # normalization condition

    for j in range(i+1,6):
        r.append(0)  # machines only repaired when no machines left working (lower-triangular matrix)

    p.append(r)

# for q in p:
#     print(q)

for line in np.array(p):
    print("  ".join(map(str, line)))
```

### part b

```python
p_restricted = []

for row in p[1:]:
    p_restricted.append(row[1:])
```

```python
p_restricted_inv = np.linalg.inv(np.identity(5)-p_restricted)   # (1-p_r)^-1

for line in p_restricted_inv:
    print("  ".join(map(str, line)))
```

```python
t = np.array([1,1,1,1,1])

# expected time until no machines are working starting from 1, 2, 3, 4, or 5
# the last element of the resulting vector is the expected time no 
# # newly installed machines are working
np.matvec(p_restricted_inv, t) 
```
