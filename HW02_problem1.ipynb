```python
import matplotlib.pyplot as plt
import regex
import numpy as np
import random   # allows random number generation
random.seed(1234)

n = 100

x_0 = 10
gamma = 4

def pdf(x):
  return (gamma - 1) * x_0**(gamma - 1) * x**(-gamma) # theoretical PDF
x = np.linspace(0,60,1000) # x-values to overlay theoretical PDF on plots

U_all = []  # list to store U(0,1) random variables
X_all = []  # list to store random variables sampled from X

for i in range(1,4):
  U_all.append([random.uniform(0,1) for _ in range(n**i)]) # generate n^i uniform random numbers
  X_all.append([x_0/u**(1/(gamma - 1)) for u in U_all[i-1]]) # sample from X
```

```python
plt.hist(X_all[0], density = True, range =(0,60))
plt.title(f'N={n}')
plt.ylim([0,0.08])
plt.xlim([0,60])
plt.plot(x,pdf(x),'red')
plt.xlabel(r'X ~ $(\gamma-1){x_0}^{\gamma-1}x^{-\gamma}$')

plt.show()
```
