```python
import scipy
import matplotlib.pyplot as plt
import regex
import numpy as np
import random   # allows random number generation
random.seed(1234)

n = [100, 1000, 10000]

x_0 = 10
gamma = 4

def pdf(x):
  return (gamma - 1) * x_0**(gamma - 1) * x**(-gamma) # theoretical PDF

x = np.linspace(10,60,1000) # x-values to overlay theoretical PDF on plots

# U_all = []  # list to store U(0,1) random variables
X_all = []  # list to store random variables sampled from X

for i in range(1,4):
  # U_all.append([random.uniform(0,1) for _ in range(n**i)]) # generate n^i uniform random numbers
  X_all.append([x_0/random.uniform(0,1)**(1/(gamma - 1)) for _ in range(n[i-1])]) # sample from X

print(len(X_all))
```

```python
plt.hist(X_all[0], density = True, range =(0,60))
plt.title(f'N={n[0]}')
plt.ylim([0,0.08])
plt.xlim([0,60])
plt.plot(x,pdf(x),'red', label = r'$f(x)$')
plt.xlabel(r'$F^{-1}(U)$')
plt.ylabel(r'$U\sim U(0,1)$')
plt.legend()
plt.show()
```
